```python
#identify outliers
import numpy as np
#calculate the quartiles and IQR
q25, q50,q75 = np.percentile(Student_Grades['Grade'], [25, 50,75])
iqr = q75 - q25
#calculate the minimum and maximum values without outliers
min_grade = q25 - 1.5 * iqr
max_grade = q75 + 1.5 * iqr
#show the calculated values
min_grade,q25,q50,q75,max_grade
```

Standard Deviation

```python
import numpy as np
#Calculate the mean and standard deviation
mean = np.mean(Student_Grades['Grade'])
sd = np.std(Student_Grades['Grade'])
#show the calculated values
mean, sd
```

```python
#Identify points that are more than 3 standard deviations away from the mean 
outliers = Student_Grades[(Student_Grades['Grade'] < mean - 3 * sd) | (Student_Grades['Grade'] > mean + 3 * sd)]
outliers
```
