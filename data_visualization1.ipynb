```python
sales=[0,5,155,0,510,0,1207,616,317,325]
month=["Jan","Feb","March",'April',"May","June","July","August","Sept","Oct"]
sales_month=pd.Series(sales,index=month,name="Sales")
```

```python
sales_month
```

```python
sales_month["Jan":"August"]
```

```python
sales_month.iloc[0:8]
```

```python
sales_month.loc["Jan":"August"]
```

```python
#Resetting the index values
```

```python
sales_month.reset_index()
```

```python
sales_month.reset_index(drop=True)
```

```python
sales_month[sales_month>510]
```

```python
sales_month[(sales_month>510) | (sales_month==325)]                #we can use & and | for multile filters
```

```python
sales_month[(sales_month<=616) & (sales_month>0)]
```

```python
sales_month.sort_values() #by default ascending order values
```

```python
sales_month.sort_values(ascending=False)   #Decending order  values
```

```python
sales_month
```

```python
sales_month.sort_values(ascending=False,inplace=True)
sales_month
```

```python
sales_month.sort_index()                    #index ascending
```

```python
sales_month.sort_index(ascending=False)             #Index descending
```
